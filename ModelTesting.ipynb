```python
import pandas as pd

LSTM = pd.read_csv('./Downloads/LSTMsubsample_rmses.csv')
GAN= pd.read_csv('./Downloads/GANsubsample_rmses.csv')
WGAN = pd.read_csv('./Downloads/WGANsubsample_rmses.csv')
MHGAN = pd.read_csv('./Downloads/MHGANsubsample_rmses.csv')
```

LSTM vs GAN

```python
from scipy import stats


t_stat, p_value = stats.ttest_ind(LSTM, GAN)

print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')
```

LSTM vs WGAN

```python
from scipy import stats


t_stat, p_value = stats.ttest_ind(LSTM, WGAN)

print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')
```
